### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
facturacion_t1 = tienda["Precio"].sum()
facturacion_t2 = tienda2["Precio"].sum()
facturacion_t3 = tienda3["Precio"].sum()
facturacion_t4 = tienda4["Precio"].sum()

print("Tienda 1:", facturacion_t1)
print("Tienda 2:", facturacion_t2)
print("Tienda 3:", facturacion_t3)
print("Tienda 4:", facturacion_t4)

facturacion = {
    "Tienda 1": facturacion_t1,
    "Tienda 2": facturacion_t2,
    "Tienda 3": facturacion_t3,
    "Tienda 4": facturacion_t4
}

tienda_mayor = max(facturacion, key=facturacion.get)
print("Tienda con mayor facturación:", tienda_mayor)
print("Monto:", f"{facturacion[tienda_mayor]:,.0f}")
```

# 2. Ventas por categoría

```python
#TIENDA 1
cat1 = tienda.groupby("Categoría del Producto").size().sort_values(ascending=False)
print("Tienda 1 - Ventas por categoría (Top 5)")
display(cat1.head(5))
```

```python
#TIENDA 2
cat2 = tienda2.groupby("Categoría del Producto").size().sort_values(ascending=False)
print("Tienda 2 - Ventas por categoría (Top 5)")
display(cat2.head(5))
```

```python
# TIENDA 3
cat3 = tienda3.groupby("Categoría del Producto").size().sort_values(ascending=False)
print("Tienda 3 - Ventas por categoría (Top 5)")
display(cat3.head(5))
```

```python
# TIENDA 4
cat4 = tienda4.groupby("Categoría del Producto").size().sort_values(ascending=False)
print("Tienda 4 - Ventas por categoría (Top 5)")
display(cat4.head(5))
```

```python
print("Resumen: categoría más popular por tienda")
print("Tienda 1:", cat1.index[0], "-", int(cat1.iloc[0]))
print("Tienda 2:", cat2.index[0], "-", int(cat2.iloc[0]))
print("Tienda 3:", cat3.index[0], "-", int(cat3.iloc[0]))
print("Tienda 4:", cat4.index[0], "-", int(cat4.iloc[0]))
```

# 3. Calificación promedio de la tienda


```python
#TIENDA 1,2,3 Y 4
prom_t1 = tienda["Calificación"].mean()
prom_t2 = tienda2["Calificación"].mean()
prom_t3 = tienda3["Calificación"].mean()
prom_t4 = tienda4["Calificación"].mean()

print("Tienda 1 - Calificación promedio:", round(prom_t1, 2))
print("Tienda 2 - Calificación promedio:", round(prom_t2, 2))
print("Tienda 3 - Calificación promedio:", round(prom_t3, 2))
print("Tienda 4 - Calificación promedio:", round(prom_t4, 2))

calif_prom = {
    "Tienda 1": prom_t1,
    "Tienda 2": prom_t2,
    "Tienda 3": prom_t3,
    "Tienda 4": prom_t4
}

mejor = max(calif_prom, key=calif_prom.get)
peor = min(calif_prom, key=calif_prom.get)

print("\nMejor satisfacción:", mejor, "-", round(calif_prom[mejor], 2))
print("Menor satisfacción:", peor, "-", round(calif_prom[peor], 2))
```

# 4. Productos más y menos vendidos
