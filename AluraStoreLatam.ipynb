### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
facturacion_t1 = tienda["Precio"].sum()
facturacion_t2 = tienda2["Precio"].sum()
facturacion_t3 = tienda3["Precio"].sum()
facturacion_t4 = tienda4["Precio"].sum()

print("Tienda 1:", facturacion_t1)
print("Tienda 2:", facturacion_t2)
print("Tienda 3:", facturacion_t3)
print("Tienda 4:", facturacion_t4)

facturacion = {
    "Tienda 1": facturacion_t1,
    "Tienda 2": facturacion_t2,
    "Tienda 3": facturacion_t3,
    "Tienda 4": facturacion_t4
}

tienda_mayor = max(facturacion, key=facturacion.get)
print("Tienda con mayor facturación:", tienda_mayor)
print("Monto:", f"{facturacion[tienda_mayor]:,.0f}")
```

# 2. Ventas por categoría

```python
#TIENDA 1
cat1 = tienda.groupby("Categoría del Producto").size().sort_values(ascending=False)
print("Tienda 1 - Ventas por categoría (Top 5)")
display(cat1.head(5))
```

```python
#TIENDA 2
cat2 = tienda2.groupby("Categoría del Producto").size().sort_values(ascending=False)
print("Tienda 2 - Ventas por categoría (Top 5)")
display(cat2.head(5))
```

```python
# TIENDA 3
cat3 = tienda3.groupby("Categoría del Producto").size().sort_values(ascending=False)
print("Tienda 3 - Ventas por categoría (Top 5)")
display(cat3.head(5))
```

```python
# TIENDA 4
cat4 = tienda4.groupby("Categoría del Producto").size().sort_values(ascending=False)
print("Tienda 4 - Ventas por categoría (Top 5)")
display(cat4.head(5))
```

```python
print("Resumen: categoría más popular por tienda")
print("Tienda 1:", cat1.index[0], "-", int(cat1.iloc[0]))
print("Tienda 2:", cat2.index[0], "-", int(cat2.iloc[0]))
print("Tienda 3:", cat3.index[0], "-", int(cat3.iloc[0]))
print("Tienda 4:", cat4.index[0], "-", int(cat4.iloc[0]))
```

# 3. Calificación promedio de la tienda


```python
#TIENDA 1,2,3 Y 4
prom_t1 = tienda["Calificación"].mean()
prom_t2 = tienda2["Calificación"].mean()
prom_t3 = tienda3["Calificación"].mean()
prom_t4 = tienda4["Calificación"].mean()

print("Tienda 1 - Calificación promedio:", round(prom_t1, 2))
print("Tienda 2 - Calificación promedio:", round(prom_t2, 2))
print("Tienda 3 - Calificación promedio:", round(prom_t3, 2))
print("Tienda 4 - Calificación promedio:", round(prom_t4, 2))

calif_prom = {
    "Tienda 1": prom_t1,
    "Tienda 2": prom_t2,
    "Tienda 3": prom_t3,
    "Tienda 4": prom_t4
}

mejor = max(calif_prom, key=calif_prom.get)
peor = min(calif_prom, key=calif_prom.get)

print("\nMejor satisfacción:", mejor, "-", round(calif_prom[mejor], 2))
print("Menor satisfacción:", peor, "-", round(calif_prom[peor], 2))
```

# 5. Envío promedio por tienda

```python
#TIENDA 1,2,3 Y 4 (ENVIOS POR TIENDA)
envio_t1 = tienda["Costo de envío"].mean()
envio_t2 = tienda2["Costo de envío"].mean()
envio_t3 = tienda3["Costo de envío"].mean()
envio_t4 = tienda4["Costo de envío"].mean()

print("Tienda 1 - Envío promedio:", round(envio_t1, 2))
print("Tienda 2 - Envío promedio:", round(envio_t2, 2))
print("Tienda 3 - Envío promedio:", round(envio_t3, 2))
print("Tienda 4 - Envío promedio:", round(envio_t4, 2))

envios = {
    "Tienda 1": envio_t1,
    "Tienda 2": envio_t2,
    "Tienda 3": envio_t3,
    "Tienda 4": envio_t4
}

mayor = max(envios, key=envios.get)
menor = min(envios, key=envios.get)

print("\nEnvío promedio más alto:", mayor, "-", round(envios[mayor], 2))
print("Envío promedio más bajo:", menor, "-", round(envios[menor], 2))
```

#GRAFICOS

####Gráfico de barras — Facturación total por tienda

```python
import pandas as pd
import matplotlib.pyplot as plt

tiendas_n = ["Tienda 1", "Tienda 2", "Tienda 3", "Tienda 4"]
fact = [
    tienda["Precio"].sum(),
    tienda2["Precio"].sum(),
    tienda3["Precio"].sum(),
    tienda4["Precio"].sum()
]

# Para que se lea mejor en pantalla (millones)
fact_m = [x / 1_000_000 for x in fact]

plt.figure(figsize=(8,4))
plt.bar(tiendas_n, fact_m)
plt.title("Facturación total por tienda")
plt.ylabel("Facturación (millones)")
plt.xlabel("Tienda")
plt.show()
```

###Gráfico circular — Categorías más populares (por ventas) en una tienda

```python
cat = tienda["Categoría del Producto"].value_counts()

top5 = cat.head(5)
otros = cat.iloc[5:].sum()

labels = list(top5.index) + ["Otros"]
sizes = list(top5.values) + [otros]

plt.figure(figsize=(7,7))
plt.pie(sizes, labels=labels, autopct="%1.1f%%")
plt.title("Distribución de ventas por categoría - Tienda 1")
plt.show()
```

```python
vc = tienda["Categoría del Producto"].value_counts()

print("Tienda 1 - Ventas por categoría (conteo)")
display(vc)

print("\n✅ Categoría más vendida (individual):", vc.index[0], "-", int(vc.iloc[0]), "ventas")

print("\nNota sobre el gráfico:")
print("- 'Otros' NO es una categoría del dataset.")
print("- 'Otros' representa la suma de todas las categorías fuera del Top 5.")
print("- Por eso 'Otros' puede ser mayor que 'Muebles', aunque 'Muebles' sea la categoría individual con más ventas.")

#Interpretación del gráfico (Tienda 1):
#Aunque en el gráfico “Otros” aparece como el segmento más grande, esto se debe a que “Otros” agrupa varias categorías que no están en el Top 5. La categoría individual con mayor número de ventas sigue siendo Muebles, como se confirma con el conteo de ventas por categoría (value_counts()).
```

###Envío promedio vs Calificación promedio

```python

envios = [
    tienda["Costo de envío"].mean(),
    tienda2["Costo de envío"].mean(),
    tienda3["Costo de envío"].mean(),
    tienda4["Costo de envío"].mean()
]

calif = [
    tienda["Calificación"].mean(),
    tienda2["Calificación"].mean(),
    tienda3["Calificación"].mean(),
    tienda4["Calificación"].mean()
]

plt.figure(figsize=(7,4))
plt.scatter(envios, calif)
plt.title("Envío promedio vs Calificación promedio")
plt.xlabel("Costo de envío promedio")
plt.ylabel("Calificación promedio")

# Etiquetas para identificar cada punto
for i in range(len(tiendas_n)):
    plt.annotate(tiendas_n[i], (envios[i], calif[i]))

plt.show()
```

#**1)Introducción**


El propósito de este análisis es apoyar al Sr. Juan en la decisión de qué tienda de Alura Store debería vender para financiar un nuevo emprendimiento. Para tomar una decisión basada en datos, se analizaron las 4 tiendas considerando indicadores de desempeño comercial y experiencia del cliente: facturación total, categorías más vendidas, productos más y menos vendidos, calificación promedio y costo de envío promedio.
Al final, se complementó el análisis con visualizaciones para identificar patrones y comparar tiendas de forma más clara.

#**2) Resultados del analisis**

### a) Facturación total (ingresos)

#### Se calculó la facturación total sumando la columna Precio en cada tienda. Con esto se puede ver cuál tienda genera más ingresos y cuál aporta menos al negocio.
#### En los resultados, se observa que la Tienda 1 es la que más factura, mientras que la Tienda 4 es la que menos factura.
Este punto es relevante porque, si el objetivo es vender una tienda, normalmente conviene vender la que menos contribuye en ingresos.

#b) Categorías más populares (ventas por categoría)


#### Se agruparon las ventas por Categoría del Producto y se contó cuántas ventas tuvo cada categoría en cada tienda. Esto permite identificar en qué tipo de productos se concentra la demanda.
#### En general, las tiendas tienen un comportamiento similar y Muebles aparece como la categoría individual con mayor número de ventas en las cuatro tiendas (por conteo de ventas).
#### En el gráfico circular de categorías (Top 5 + “Otros”), se debe considerar que “Otros” no es una categoría real, sino la suma de las categorías fuera del Top 5. Por eso “Otros” puede verse más grande que Muebles, aunque Muebles siga siendo la categoría No. 1 individualmente.

# c) Calificación promedio (satisfacción del cliente)

#### Para medir satisfacción, se calculó el promedio de Calificación por tienda. Este indicador ayuda a comparar la percepción del cliente entre tiendas.En los resultados, la Tienda 3 presenta la mejor calificación promedio, y la Tienda 1 la más baja, aunque las diferencias no son extremas (las calificaciones están cercanas a 4).Esto sugiere que Tienda 3 tiene una mejor experiencia percibida por el cliente, mientras que Tienda 1 podría tener oportunidades de mejora en satisfacción.

# d) Productos más vendidos y menos vendidos

#### Se identificaron los productos más vendidos y menos vendidos usando el conteo de apariciones en la columna Producto (value_counts()), lo cual es válido porque cada fila representa una venta.
#### Para los menos vendidos, es común encontrar varios productos empatados con ventas mínimas (por ejemplo 1 venta), por lo que mostrar un Top 10 de menos vendidos es más informativo que elegir solo uno.

#### Este análisis sirve para entender qué productos “empujan” las ventas y cuáles no están funcionando en cada tienda.

# e) Costo de envío promedio

#### Se calculó el promedio de Costo de envío por tienda. Aunque el envío lo paga el cliente, este valor es útil porque puede influir en la experiencia de compra y en la percepción del servicio.

#### En el gráfico de dispersión (envío promedio vs calificación promedio) se observa un punto relevante:


*   Tienda 1 presenta el envío promedio más alto y, al mismo tiempo, la calificación promedio más baja.
Esto es una señal negativa, ya que el cliente paga más envío y la satisfacción no es la mejor.
*   Tienda 3 logra la mejor calificación con un envío intermedio.


*   Tienda 4 tiene el envío promedio más bajo, pero no es la mejor evaluada, lo que indica que la satisfacción depende de más factores además del costo de envío.

#**3) Conclusión y recomendación**

Con base en los indicadores analizados y las visualizaciones generadas, la recomendación es que el Sr. Juan venda la Tienda 4.

Motivos principales:

*   Es la tienda con menor facturación total, por lo que aporta menos ingresos al negocio comparado con las demás.
*   Aunque mantiene categorías populares similares a las otras tiendas, no muestra una ventaja clara que compense su menor desempeño en ingresos.
*   Su calificación promedio no es la mejor, por lo que tampoco destaca por satisfacción como para justificar mantenerla por experiencia del cliente.
*   Tiene un envío promedio bajo, pero eso no se traduce en mejor satisfacción ni en un mejor rendimiento general.


En resumen, la Tienda 4 es la opción más razonable para vender porque es la que muestra el desempeño más débil a nivel global, principalmente en el indicador más importante para una decisión de venta: la facturación.
